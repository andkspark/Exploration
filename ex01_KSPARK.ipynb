```python
#import warnings
#warnings.filterwarnings("ignore")
#데이터셋 로드
from sklearn.datasets import load_digits
from sklearn.datasets import load_wine
from sklearn.datasets import load_breast_cancer

#모델
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression


#공통으로 사용되는 모듈 내 함수
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
```

```python
#Feature Data 지정하기
#Label Data 지정하기
#Target Names 출력해 보기
#데이터 Describe 해 보기

#데이터 로드, 손글씨
digits = load_digits()
digits_data = digits.data
digits_label = digits.target

#정보를 일단 알아보죠
print(digits.DESCR)
print()
```

```python
#필요한 정보 자세히보기
print("손글씨 내부 객체 정보 ")
print(dir(digits))
print()
print("손글씨 key : ", digits.keys())
print()
print("손글씨 데이터의 행렬 형상 : ", digits_data.shape)
print()
print("손글씨 데이터의 생김새")
print(digits_data[0])
print()
print("손글씨 타겟정보(분류되는 라벨)의 형상 : ", digits_label.shape)
print()
print("손글씨 타겟정보의 생김새")
print(digits_label)
print()
print("손글씨 데이터셋의 피쳐 이름")
print(digits.feature_names)
```

```python
#정답지 준비
digits_df = pd.DataFrame(data = digits_data, columns = digits.feature_names)
#정답행렬 추가 후, 라벨 정보를 넣어주기
digits_df['label'] = digits.target
digits_df
#맨 끝에 라벨 정보가 잘 붙었음
```

```python
#학습 비교를 위해 트레인과 테스트셋 비율을 다양하게 준비
#0.15 0.20 0.25
#랜덤시드는 동일하게 씀

X_train_digits_015, X_test_digits_015, y_train_digits_015, y_test_digits_015 = train_test_split(digits_data, digits_label, test_size=0.15, random_state=7)
X_train_digits_020, X_test_digits_020, y_train_digits_020, y_test_digits_020 = train_test_split(digits_data, digits_label, test_size=0.2, random_state=7)
X_train_digits_025, X_test_digits_025, y_train_digits_025, y_test_digits_025 = train_test_split(digits_data, digits_label, test_size=0.25, random_state=7)

print('X_train digits 015 개수: ', len(X_train_digits_015),', X_test digits 015 개수: ', len(X_test_digits_015))
print('X_train digits 020 개수: ', len(X_train_digits_020),', X_test digits 020 개수: ', len(X_test_digits_020))
print('X_train digits 025 개수: ', len(X_train_digits_025),', X_test digits 025 개수: ', len(X_test_digits_025))

#테스트 데이터 라벨이 잘 섞였는지 체크
y_test_digits_020
```

```python
#결정트리 학습시키기
#마찬가지로 각 비율로 나눠진 데이터셋에 대해, 랜덤 시드는 같은 수를 씀
decision_tree_digits_015 = DecisionTreeClassifier(random_state = 32)
decision_tree_digits_020 = DecisionTreeClassifier(random_state = 32)
decision_tree_digits_025 = DecisionTreeClassifier(random_state = 32)

#1줄 학습, 1줄 테스트, 1줄 정확도 분석
#데이터의 85%를 학습에 사용한 경우
decision_tree_digits_015.fit(X_train_digits_015, y_train_digits_015)
y_pred_digits_DT_015 = decision_tree_digits_015.predict(X_test_digits_015)
accuracy_digits_DT_015 = accuracy_score(y_test_digits_015, y_pred_digits_DT_015)

#같은 순서로 데이터의 80%를 학습에 사용한 경우
decision_tree_digits_020.fit(X_train_digits_020, y_train_digits_020)
y_pred_digits_DT_020 = decision_tree_digits_020.predict(X_test_digits_020)
accuracy_digits_DT_020 = accuracy_score(y_test_digits_020, y_pred_digits_DT_020)

#같은 순서로 데이터의 75%를 학습에 사용한 경우
decision_tree_digits_025.fit(X_train_digits_025, y_train_digits_025)
y_pred_digits_DT_025 = decision_tree_digits_025.predict(X_test_digits_025)
accuracy_digits_DT_025 = accuracy_score(y_test_digits_025, y_pred_digits_DT_025)
```

```python

#정확도 분석
print("Decision tree 데이터 85%를 학습, 15%를 테스트에 사용 : ", accuracy_digits_DT_015)
print("Decision tree 데이터 80%를 학습, 20%를 테스트에 사용 : ", accuracy_digits_DT_020)
print("Decision tree 데이터 75%를 학습, 25%를 테스트에 사용 : ", accuracy_digits_DT_025)
print()

#컨퓨전 매트릭스로 보자
print("컨퓨전 매트릭스 : Decision tree 데이터 85%를 학습, 15%를 테스트에 사용")
print(classification_report(y_test_digits_015, y_pred_digits_DT_015))
print()

print("컨퓨전 매트릭스 : Decision tree 데이터 80%를 학습, 20%를 테스트에 사용")
print(classification_report(y_test_digits_020, y_pred_digits_DT_020))
print()

print("컨퓨전 매트릭스 : Decision tree 데이터 75%를 학습, 25%를 테스트에 사용")
print(classification_report(y_test_digits_025, y_pred_digits_DT_025))
print()
```

```python
#짧은 분석 - f1 score가 클래스에 따라 널이 뛴다
#85% 트레인 - 15% 테스트에 대해, 일반적으로 훈련에 소비된 데이터 양이 많을수록 훈련이 잘 되었다고 유추해야 하지만 성공률이 살짝 낮다
#이는 테스트를 위한 데이터 표본이 적어서, 잘 보여주지 못한 것으로 예측됨
```

```python
logistic_model_digits_015 = LogisticRegression(max_iter=3000)
logistic_model_digits_020 = LogisticRegression(max_iter=3000)
logistic_model_digits_025 = LogisticRegression(max_iter=3000)

#컨버전스 워닝이 떠서 max iteration 제한
#참고 : https://stackoverflow.com/questions/62658215/convergencewarning-lbfgs-failed-to-converge-status-1-stop-total-no-of-iter

logistic_model_digits_015.fit(X_train_digits_015, y_train_digits_015)
y_pred_digits_LR_015 = logistic_model_digits_015.predict(X_test_digits_015)
accuracy_digits_LR_015 = accuracy_score(y_test_digits_015, y_pred_digits_LR_015)

logistic_model_digits_020.fit(X_train_digits_020, y_train_digits_020)
y_pred_digits_LR_020 =logistic_model_digits_020.predict(X_test_digits_020)
accuracy_digits_LR_020 = accuracy_score(y_test_digits_020, y_pred_digits_LR_020)

logistic_model_digits_025.fit(X_train_digits_025, y_train_digits_025)
y_pred_digits_LR_025 = logistic_model_digits_025.predict(X_test_digits_025)
accuracy_digits_LR_025 = accuracy_score(y_test_digits_025, y_pred_digits_LR_025)
```
